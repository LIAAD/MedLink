```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('./data/filtered_case_reports.csv')
df
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from unidecode import unidecode
from sklearn.metrics.pairwise import cosine_similarity



vectorizer = TfidfVectorizer(tokenizer=lambda x: x, lowercase=False)
tfidf_matrix = vectorizer.fit_transform(df['processed_keywords'])

#vetorizar com tf-idf ?
vectorizer = TfidfVectorizer(tokenizer=lambda x: x, lowercase=False)
tfidf_matrix = vectorizer.fit_transform(df['processed_keywords'])

#criar clusters

num_clusters = 11  # Define number of clusters
kmeans = KMeans(n_clusters=num_clusters)
df['cluster'] = kmeans.fit_predict(tfidf_matrix)

#dbscan = DBSCAN(eps=0.5, min_samples=10)
#df['cluster'] = dbscan.fit_predict(tfidf_matrix.toarray()) 
#print(df_non_nan[['title', 'keywords', 'cluster']])

cluster_counts = df['cluster'].value_counts()
print("\nCluster counts:\n", cluster_counts)
```

```python
# Function to show titles for a specific cluster
def show_titles_for_cluster(cluster_number):
    titles = df[df['cluster'] == cluster_number]['title']
    keywords = df[df['cluster'] == cluster_number]['processed_keywords']
    return titles, keywords

# Example: Show titles for cluster  4, 8 
cluster_number = 7
titles, keywords = show_titles_for_cluster(cluster_number)

print(f"Titles in Cluster {cluster_number}:")
for title, keyword in zip(titles, keywords):
    print(title, keyword)
```
