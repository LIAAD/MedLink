### Load Data

```python
import pandas as pd
filtered_df = pd.read_csv('data/filtered_case_reports.csv')
abstracts = filtered_df['abstract'].tolist()
keywords = filtered_df['keywords'].tolist()
filtered_df
```

##### Load generated reports (queries)

```python
queries_df = pd.read_csv('./data/generated_medical_reports.csv')
#add id row to df
ids = [192, 111, 157, 726, 824, 287, 859, 322, 38, 85]
queries_df['id'] = ids
queries_df
```

### Load Bi-Encoder

```python
#test the model on inference given a an abstract and the list of abstracts
from sentence_transformers import SentenceTransformer, util
import torch
# Load the model
model = SentenceTransformer("lfcc/medlink-bi-encoder") 

#Calculate the abstract embeddings - This only need to be done once
embeddings = model.encode(abstracts, convert_to_tensor=True)
```

### Retrieve documents given a query

```python
query_text = queries_df["Texto Clinico"][1]

# Encode the query
query_embedding = model.encode(query_text, convert_to_tensor=True)

# Calculate the similarity between the query and the abstracts
cosine_scores = util.pytorch_cos_sim(query_embedding, embeddings)
retrieval_results = torch.topk(cosine_scores, k=10)

for score, idx in zip(retrieval_results.values[0], retrieval_results.indices[0]):

    print(f"Sentence: {abstracts[idx]}\nKeywords: {keywords[idx]} \nSimilarity Score: {score.item():.4f}\n")
```

```python
retrieval_results
```

```python
retrieved_abstracts = [abstracts[idx] for idx in retrieval_results.indices[0]]
```

### Load Cross Encoder

```python
from sentence_transformers import CrossEncoder
cross_model = CrossEncoder("lfcc/medlink-cross-encoder", num_labels=1, max_length=512)
```

### Rerank Documents

```python

reranking_results = cross_model.rank(query_text, retrieved_abstracts)

# Print the scores
#print("Query:", query)
for rank in reranking_results:
    abstract_idx = rank['corpus_id']  # Get the index of the ranked abstract
    print(f"Score: {rank['score']:.2f}")
    print(f"Abstract: {retrieved_abstracts[abstract_idx]}")
    print("=" * 80)  # Separator for better readability
```

```python
#get reranked abstracts with the cross encoder in the filtered_df
reranked_abstracts = []

for rank in reranking_results:
    retrieval_idx = rank['corpus_id']
    #print(retrieval_results.indices[0][retrieval_idx])
    abstract_idx = retrieval_results.indices[0][retrieval_idx].cpu().numpy()
    # add score to 
    #get entries from the filtered_df
    paper = filtered_df.iloc[abstract_idx].copy()
    paper['score'] = rank['score']
    reranked_abstracts.append(paper)
reranked_abstracts[0]
```
